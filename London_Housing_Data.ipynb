```python
import pandas as pd
data = pd.read_csv(r"C:/Users/MKRR/Desktop/Dataset/London Housing Data.csv")
data
```

```python
#(1) Convert the Datatypes of Date column to Date-Time format.
data.dtypes
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.dtypes
```

```python
#(2) Add a new column year which contains only year
data['year'] = data['date'].dt.year
```

```python
data
```

```python
# (3) Add a new column "month" as 2nd column in the dataframe, which contains only month
data.insert(1, "mnth", data['date'].dt.month)
```

```python
data
```

```python
#(4) Remove the columns year and month
data.drop( ['year', 'mnth'], axis =1, inplace = True )
```
